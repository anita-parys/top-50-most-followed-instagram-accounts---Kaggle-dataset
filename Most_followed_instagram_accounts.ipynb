```python
#https://www.kaggle.com/datasets/hassanshehzadk/top-50-most-followed-instagram-accounts?resource=download
import pandas as pd
df = pd.read_csv("top.csv",encoding='unicode_escape')
```

```python
df.head(10)
```

##which country has the most instagrammers from top 50

```python
country=df.copy()
country=country.groupby(country['Country/Continent']).count()
country = country.sort_values("Owner",ascending=False)
country = country.drop(columns = ["Rank","Username", "Followers (millions)", "Profession/Activity"])
country=country.rename(columns={"Owner": "Number of the most followed accounts"})
country.head()
```

##which profession/activity is the most popular in top 50 most followed instagram accounts

```python
followers=df.copy()
followers=followers.groupby(followers['Profession/Activity']).count()
followers = followers.sort_values("Rank", ascending = False)
followers.head(10)
```

###Which music artist has the most followers

```python
musician = df.loc[df["Profession/Activity"].str.contains("Musician")]
musician.loc[musician["Followers (millions)"].idxmax()]
```

### Top 5 most followed music artists on instagram

```python
musician.head()
```

###Which athletes has the most followers

```python
sport= ["Footballer", "Cricketer","Basketball player"]
athletes = df.loc[df["Profession/Activity"].isin(sport)]
athletes.loc[athletes["Followers (millions)"].idxmax()]
```

###Top 5 most followed athletes on instagram

```python
athletes.head()
```

### Top 5 most followed tv personality and model on instagram

```python
tv = df.loc[df["Profession/Activity"].str.contains("Television personality|Model")]
tv.head(5)
```

### Top 5 Most followed actor and actress on instagram

```python
actor = df.loc[df["Profession/Activity"].str.contains("Actor|Actress")]
actor.head()
```

###Who has the most followers from Top 5: musician, athletes, television personality or actors? 

```python
musician_followers = musician.head()["Followers (millions)"].sum()
athletes_followers = athletes.head()["Followers (millions)"].sum()
tv_followers = tv.head()["Followers (millions)"].sum()
actor_followers = actor.head()["Followers (millions)"].sum()

data = {'Followers (millions)': [musician_followers,athletes_followers,tv_followers,actor_followers]}
df1 = pd.DataFrame(data,index = ['Musician', 'Athletes','Television personality', 'Actors'])
df1.sort_values("Followers (millions)", ascending = False)
```
